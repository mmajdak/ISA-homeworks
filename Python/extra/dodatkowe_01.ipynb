```python
# Napisz funkcję create_tower, która jako input przyjmuje integera większego od 0
# i zwraca listę, która tworzy "piramidę" z tyloma piętrami:
# Przykładowe rozwiązanie:
# In: create_tower(3)
# Out:
# [
#  '  *  ', 
#  ' *** ', 
#  '*****'
# ]
```

```python
def create_tower(x):
    for i in range (0,x):
        txt=i*'*'+'*'+i*'*'
        print(txt.center(x*2))

x = int(input('Jaka ma byc wysokosc choinki? '))
print('\n\n')
create_tower(x)
```

```python
import pandas as pd

filename = "all_seasons.csv"
data_frame = pd.read_csv(filename)
data_frame
```

```python
# a) Czy jest jakikolwiek zawodnik spoza USA, który przed NBA uczęszczał do college'u Kentucky?
```

```python
# b) Ile wynosi średni wzrost zawodnika w calach??
#    Czy od pierwszego sezonu w tym zbiorze do teraz powiększył się czy zmniejszył?
```

```python
import numpy as np
import matplotlib.pyplot as plt

frame_height = data_frame[['season','player_height']]
group_seasons = frame_height.groupby('season')
height_per_season = group_seasons.mean()

a = height_per_season['player_height'].to_numpy()

print(f"Średni wzrost zawodnika w calach wynosi {round(a.mean()*0.3937,2)}.\n")
if a[0] > a[-1]:
    komunikat = 'zmniejszył się'
elif a[0] < a[-1]:
    komunikat = 'zwiększył się'
else:
    komunikat = 'nie zmienił się'
print(f"Od pierwszego badanego sezonu średni wzrost zawodników {komunikat}.\nW pierwszym sezonie był on na poziomie {round(a[0]*0.3937,2)} cali, a w ostatnim było to {round(a[-1]*0.3937,2)} cali.")
```

```python
# c) Kto w badanym okresie rzucił łącznie najwięcej punktów?
#    Czy ta osoba miała też największą średnią punktów na sezon?
```

```python
# Z danych niestety nie ma możliwości wyciągnąć liczbę punktów dla całego danego okresu.
# W kolumnie pts są jedynie średnie punktowe za poszczególny sezon.
# Sumowanie średnich sezon jest bezsensu, więc ze średnich wyciągnałem kolejną średnią - tym razem dla całego okresu.
# W badanym okresie najlepszy był LeBron James, który zdobywał przeciętnie 26,94 punktów na mecz

players_pts = data_frame[['player_name','pts','season']]
avg_player = players_pts.groupby('player_name')
avg_total = avg_player['pts'].mean().sort_values(ascending=False).head(1)
name_pts = avg_total.to_dict()
print(name_pts)
```

```python
# Z poszczególnych sezonów wyciągnąłem strzelców z najwyższą średnią i sprawdziłem ile razy był wśród nich Lebron.

uniq_season = data_frame['season'].unique().tolist()
lebron = 0
for x in range(0,len(uniq_season)):
    a = data_frame.loc[data_frame['season'] == uniq_season[x]]['pts'].idxmax()
    b = data_frame.loc[[a],['player_name','pts']].to_dict()
    if b['player_name'][a] == 'LeBron James':
        lebron += 1

print(f"LeBron James miał {lebron} raz(y) najlepszą średnią punktową za poszczególny sezon.")
```

```python
# d) W której rundzie draftu największy udział procentowy mieli zawodnicy ważący więcej niż 100 kg?
```

```python
# Na początku oczyszczenie danych - usunięcie duplikatów, a więc tych samych nazwisk, które powtarzają się w każdym sezonie.

draft_info = data_frame[['player_name','draft_round','player_weight']]
draft_info.drop_duplicates(subset='player_name', keep='first', inplace = True)
# draft_info

# Później usunięcie z tabeli zawodników, którzy nie byli wybierani do NBA w drafcie.
only_draft = draft_info[draft_info['draft_round'] != 'Undrafted']
only_draft['draft_round'].unique()

for x in range(1,9):
    x_round = only_draft[only_draft['draft_round'] == str(x)]
    kg = x_round[x_round['player_weight'] > 100.0]
    if len(x_round) == 0:
        percent_over100 = 0
    else: 
        percent_over100 = len(kg)/len(x_round)*100.0
        print(f"Runda {x}. udział graczy powyżej 100 kg: {percent_over100:.2f} %.")
        
# Zawodnicy ważący powyżej 100 kg mieli największy udział procentowy w 1 rundzie draft - 47.61%.
```
