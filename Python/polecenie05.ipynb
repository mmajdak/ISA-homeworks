````python
print('Kołko i krzyżyk - zaczynamy! Plansza z polami od 1 do 9.\n\n\t1 2 3\n\t4 5 6\n\t7 8 9\n')
plansza = ["-"] * 10

def rys_plansza():
    return print(f"\t{plansza[1]} {plansza[2]} {plansza[3]}\n\t{plansza[4]} {plansza[5]} {plansza[6]}\n\t{plansza[7]} {plansza[8]} {plansza[9]}\n")

def sprawdz(znak):
    komunikat = "Koniec gry - wygrywa"
    if plansza[1] == plansza[2] == plansza[3] == znak:
        return print(f"{komunikat} \'{znak}\'!"), True
    if plansza[4] == plansza[5] == plansza[6] == znak:
        return print(f"{komunikat} \'{znak}\'!"), True
    if plansza[7] == plansza[8] == plansza[9] == znak:
        return print(f"{komunikat} \'{znak}\'!"), True
    if plansza[1] == plansza[4] == plansza[7] == znak:
        return print(f"{komunikat} \'{znak}\'!"), True
    if plansza[2] == plansza[5] == plansza[8] == znak:
        return print(f"{komunikat} \'{znak}\'!"), True
    if plansza[3] == plansza[6] == plansza[9] == znak:
        return print(f"{komunikat} \'{znak}\'!"), True
    if plansza[1] == plansza[5] == plansza[9] == znak:
        return print(f"{komunikat} \'{znak}\'!"), True
    if plansza[3] == plansza[5] == plansza[7] == znak:
        return print(f"{komunikat} \'{znak}\'!"), True
    return False

def sprawdz_remis():
    if plansza[1] != "-" and plansza[2] != "-" and plansza[3] != "-" and plansza[4] != "-" and plansza[5] != "-" and plansza[6] != "-" and plansza[7] != "-" and plansza[8] != "-" and plansza[9] != "-":
        print('Koniec gry - remis')
    return False

while True:
    i=0
    while i<9:
        if i % 2 == 0: 
            o = int(input('Na którym polu postawić \'O\'? '))
            while not 0 < o < 10 or not plansza[o] == "-": 
                o = int(input('Podana pozycja jest niepoprawna. Podaj jeszcze raz, na którym polu postawić \'O\'? '))
            plansza[o] = "O"
            rys_plansza()
            if sprawdz("O"):
                break
            if sprawdz_remis():
                break                
            i += 1
        else: 
            x = int(input('Na którym polu postawić \'X\'? '))
            while not 0 < x < 10 or not plansza[x] == "-": 
                x = int(input('Podana pozycja jest niepoprawna. Podaj jeszcze raz, na którym polu postawić \'X\'? '))
            plansza[int(x)] = "X"
            rys_plansza()
            if sprawdz("X"):
                break
            if sprawdz_remis():
                break  
            i += 1
    break
````
